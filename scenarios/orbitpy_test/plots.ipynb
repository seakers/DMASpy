# Orbitpy + DMAS Integration Test

This script plots the state of a satellite performing measurements with realistic astrodynamics.

## Scenario
There exists only one agent: a satellite orbiting earth. The agent makes no measurements and has no capabilities other than station keeping. It posssess a battery with a capacity of 0.0012 kWh (4320J) and a maximum output of 10W, a solar panel capable of outputting up to 10W of power. Idling, the satellite requires 5W to maintain operations. 

## Power Tracking
Due to its orbit, the satellite enters the earth's shadow at T=667.19s and exits at time T=1726.30s. This forces the agent to use its battery power during the eclipse period. However, the small capacity of the agent's battery is not enough to maintain the satellite's operations for long enough to escape the Earth's eclipse and use its solar panels again.

At T=1522.55s the agent detects its battery going below its allowable depth of discharge and stops its power output to look for alternative sources of power. Since it has no other alternatives due to the satellite being under eclipse, it chooses to reactivate the battery power output. At T=1531.19s, the satellite's platform simulator detects a depleted battery and thus is stops ifs power output. The agent detects this and is unable to find components that can supply power to itself. Since this critical state laster for more than a timestep, the platform simulator shuts down the satellite. The agent detects this as well and turns off at T=1532s.
